```python
import plotly.graph_objs as go
```

```python
trace1 = go.Scatter(x=df_100.world_rank,y=df_100.citations,mode='lines', name='Citations', marker=dict(color='rgba(16,112,2,0.8)'), text=df.university_name)

trace2 = go.Scatter(x=df_100.world_rank,y=df_100.teaching, mode='lines+markers',name='teaching', marker=dict(color='rgba(80,23,80,0.8)'), text=df.university_name)

data=[trace1, trace2]

layout=dict(title='Citation and teaching vs World Rank of Top 100', xaxis=dict (title='Citation Rank', ticklen=5, zeroline=False))

fig=dict(data=data,layout=layout)
iplot(fig)
```

```python
df_2014 = df[df.year ==2014].iloc[:100,:]
df_2015 = df[df.year ==2015].iloc[:100,:]
df_2016 = df[df.year ==2016].iloc[:100,:]

#scatter dağılım grafiği demek

trace1 = go.Scatter(
    x=df_2014.world_rank,y=df_2014.citations, mode='markers', name="2014", marker=dict(color='rgba(255,128,2,0.8)'), text=df_2014.university_name
)
trace2 = go.Scatter(
    x=df_2015.world_rank,y=df_2015.citations, mode='markers', name="2014", marker=dict(color='rgba(0,255,200,0.8)'), text=df_2015.university_name
)
trace3 = go.Scatter(
    x=df_2016.world_rank,y=df_2016.citations, mode='markers', name="2014", marker=dict(color='rgba(255,128,255,0.8)'), text=df_2016.university_name
)
data=[trace1, trace2, trace3]

layout=dict(title='Citation vs World Rank of Top 100 uni with 2014-2015-2016', xaxis=dict (title='Citation Rank', ticklen=5, zeroline=False))

fig=dict(data=data,layout=layout)
iplot(fig)

```

```python
#bar grafikleri

import plotly.graph_objs as go

trace1 = go.Bar(
    x=df_2014.university_name,
    y=df_2014.citations,
    name="citations",
    marker=dict(color='rgba(255,128,2,0.5)'),
    line=dict(color='rgb(0,0,0)'),
    width=1.5,
    text=df_2014.country
)

trace2 = go.Bar(
    x=df_2014.university_name,
    y=df_2014.teaching,
    name="teaching",
    marker=dict(color='rgba(0,255,200,0.8)'),
    line=dict(color='rgb(0,0,0)'),
    width=1.5,
    text=df_2014.country
)

data = [trace1, trace2]

layout = go.Layout(barmode='group')

fig = go.Figure(data=data, layout=layout)

iplot(fig)
```

```python
df_2011=df.student_staff_ratio[df.year==2011]
df_2012=df.student_staff_ratio[df.year==2012]

trace=go.Histogram(
    x=df_2011,opacity=0.75, name='2011', marker=dict(color='rgba(171,50,96,0.6)')
)
trace2=go.Histogram(
    x=df_2012,opacity=0.75, name='2011', marker=dict(color='rgba(12,50,196,0.6)')
)

data=[trace,trace2]
layout=go.Layout(barmode='overlay',title='students-staff ratio in 2011 and 2012', xaxis=dict(title='students-staff-ratio'),yaxis=dict(title='Count'))
fig=go.Figure(data=data, layout=layout)
iplot(fig)
```
