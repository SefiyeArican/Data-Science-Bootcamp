```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
#kullanıcı arayüzüne çıkartacağımız için 
cut_dict = {'Fair':1, 'Good':2, 'Very Good':3, 'Premium':4, 'Ideal':5}
color_dict = {'E':1, 'I':2, 'J':3, 'H':4, 'F':5, 'G':6, 'D':7}
clarity_dict = {'SI2':1, 'SI1':2, 'VS1':3,'VS2':4, 'VVS2':5, 'VVS1':6, 'I1':7, 'IF':8}

df['color'] = df['color'].map(color_dict)
df['cut'] = df['cut'].map(cut_dict)
df['clarity'] = df['clarity'].map(clarity_dict)
```

```python
df.head()
```

```python
X = df.drop('price',axis=1)
y = df['price']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
reg = RandomForestRegressor()
reg.fit(X_train, y_train)
```

```python
reg.score(X_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(reg, X_train, y_train, cv=10).mean()
```

```python
import pickle

with open('model.pkl', 'wb') as f:  #write binary formatında => wb
    pickle.dump(reg, f)

with open('model.pkl', 'rb') as f:  #read binary formatında => rb
    rf = pickle.load(f)    

rf
```

```python
from sklearn.metrics import r2_score

preds = rf.predict(X_test)

score = r2_score(y_test, preds)
print('R2 score:', score)
```
